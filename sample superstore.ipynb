### PROFIT ANALYSIS

### Computing top categories in terms of sales from first 100 observations


```python
category_s=ss.groupby('Category').Sales.sum().nlargest(n=100)
category_p=ss.groupby('Category').Profit.sum().nlargest(n=100)
```

### PROFIT ANALYSIS OF CATEGORY

```python

plt.style.use('seaborn')
category_s.plot(kind='bar',figsize=(12,6),fontsize =14,legend=True)
category_p.plot(kind='bar',figsize=(12,6),fontsize =14,color='orange',legend=True)
plt.xlabel('Category',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title('Top Category Sales vs Profits',fontsize=15)
plt.show()
```

### PROFIT ANALYSIS OF SUB-CATEGORY

```python
category_s=ss.groupby('Sub-Category').Sales.sum().nlargest(n=100)
category_p=ss.groupby('Sub-Category').Profit.sum().nlargest(n=100)
plt.style.use('seaborn')
category_s.plot(kind='bar',figsize=(12,6),fontsize =14,legend=True)
category_p.plot(kind='bar',figsize=(12,6),fontsize =14,color='orange',legend=True)
plt.xlabel('Sub-Category',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title('Top Sub-Category Sales vs Profits',fontsize=15)
plt.show()
```

### PROFIT ANALYSIS OF REGION

```python
plt.figure(figsize=(6,6))
plt.title('Region')
plt.pie(ss['Region'].value_counts(),labels=ss['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

### PROFIT ANALYSIS OF STATES

```python
category_s=ss.groupby('State').Sales.sum().nlargest(n=100)
category_p=ss.groupby('State').Profit.sum().nlargest(n=100)
plt.style.use('seaborn')
category_s.plot(kind='bar',figsize=(12,6),fontsize =14,legend=True)
category_p.plot(kind='bar',figsize=(12,6),fontsize =14,color='orange',legend=True)
plt.xlabel('States',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title('Top States Sales vs Profits',fontsize=15)
plt.show()
```

### PROFIT ANALYSIS OF SEGMENT

```python
category_s=ss.groupby('Segment').Sales.sum().nlargest(n=100)
category_p=ss.groupby('Segment').Profit.sum().nlargest(n=100)
plt.style.use('seaborn')
category_s.plot(kind='bar',figsize=(12,6),fontsize =14,legend=True)
category_p.plot(kind='bar',figsize=(12,6),fontsize =14,color='orange',legend=True)
plt.xlabel('Segment',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title('Top Segment Sales vs Profits',fontsize=15)
plt.show()
```

### PROFIT ANALYSIS OF SHIP-MODE

```python
category_s=ss.groupby('Ship Mode').Sales.sum().nlargest(n=100)
category_p=ss.groupby('Ship Mode').Profit.sum().nlargest(n=100)
plt.style.use('seaborn')
category_s.plot(kind='bar',figsize=(12,6),fontsize =14,legend=True)
category_p.plot(kind='bar',figsize=(12,6),fontsize =14,color='orange',legend=True)
plt.xlabel('Ship Mode',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title('Top Ship Mode Sales vs Profits',fontsize=15)
plt.show()
```
